This code creates one large notebook that combines the scores for all the submetrics. Each neighborhood has a score from 1-15.

```python
totalScores = dict()
for x,y in parks.items():
        totalScores[x] = y

for x,y in dumps.items():
    if x not in totalScores:
        totalScores[x] = y
    else:
        totalScores[x] += y

for x,y in trash.items():
    if x not in totalScores:
        totalScores[x] = y
    else:
        totalScores[x] += y

sortedTotalScores = dict(sorted(totalScores.items(), key=lambda x:x[1]))
print(sortedTotalScores)
```

This creates a bargraph with the scores of the top half neighborhoods with the highest score. Data visualization allows people to easily make and see comparisons among the data. We split the data bar graph into two in order to avoid over-clustered graphs, and to allow for more clear visualization.

```python
#creating a plot for the top half of the data set
import matplotlib.pyplot as plt
firstHalf = dict(list(sortedTotalScores.items())[len(sortedTotalScores)//2:])
nhood = list(firstHalf.keys())
values = list(firstHalf.values())
plt.bar(nhood, values)
plt.xlabel('Neighborhood')
plt.ylabel('Score')
plt.title('Environmental Score : Top half')
plt.xticks(rotation = 'vertical')
plt.tight_layout()
plt.show()
```

This creates a bargraph with the scores of the bottom half neighborhoods with the lowest score.

```python
#creating a plot for the bottom half of the data set
import matplotlib.pyplot as plt
secondHalf = dict(list(sortedTotalScores.items())[:len(sortedTotalScores)//2])
nhood = list(secondHalf.keys())
values = list(secondHalf.values())
plt.bar(nhood, values)
plt.xlabel('Neighborhood')
plt.ylabel('Score')
plt.title('Environmental Score: Bottom Half')
plt.xticks(rotation = 'vertical')
plt.tight_layout()
plt.show()
```
